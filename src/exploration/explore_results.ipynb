```python
def plot_hist_comparison(title, metrics_dict, save=False):
    names = {'mlp_classification_blosum_with_allele_encoder_500_neurons_50_epochs_shuffled_64_batch_size.pt':
            'SHUFF_MLP',
            'mlp_classification_blosum_with_allele_encoder_500_neurons_50_epochs_LOGO_anch_rep_64_batch_size.pt':
            'CLUST_MLP',
            # 'mlp_classification_blosum_with_allele_encoder_500_neurons_50_epochs_marieke_cluster_64_batch_size.pt':
            # 'CLUST_MLP',
            'mlp_classification_blosum_with_allele_encoder_500_neurons_50_epochs_pseudoseq_cluster_64_batch_size.pt':
            'ALLELE_MLP',
            'mhcflurry_held_out_trained':
            'SHUFF_MHCFLURRY',
            'mhcflurry_peptide_clustered_trained':
            'PEPT_MHCFLURRY',
            'mhcflurry_allele_clustered_trained':
            'ALLELE_MHCFLURRY'
            }
    fig = go.Figure()

    for model_name in metrics_dict:
        try:
            name = names[model_name]
        except KeyError:
            name = model_name
        x = list(metrics_dict[model_name].keys())
        y = list(metrics_dict[model_name].values())
        fig.add_trace(go.Bar(
            x=x,
            y=y,
            name=name,
        ))
        fig.update_layout(
            barmode="group",
            title_text=title,
        )
    fig.show()
    if save:
        fig.write_html(save)
        
plot_hist_comparison('Shuffled-peptClust-alleleClust cnn vs mlp', best_models_metrics, save='full_comparison.html')
```
