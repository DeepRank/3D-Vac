```python
import deeprank_plots as dp
import os
import h5py
import numpy as np
import plotly.graph_objects as go
from matplotlib import pyplot as plt
import pandas as pd

metricsfile = '/projects/0/einf2380/data/pMHCI/trained_models/CNN/clustAllele_noPSSM/CnnClass4ConvKS3Lin100SoftMax/metrics.hdf5'

metricsfiles = {
    #'TEST_shuff_128B_13E' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/shuffled_noPSSM_20E128B/CnnClass4ConvKS3Lin100SoftMaxMult64/metrics.hdf5',
    #'TEST_shuff_128B_10E' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/shuffled_noPSSM_10E128B_noearlystop/CnnClass4ConvKS3Lin100SoftMaxMult64/metrics.hdf5',
    'SHUFF_CnnClass4ConvKS3Lin100SoftMax' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/shuffled_noPSSM/CnnClass4ConvKS3Lin100SoftMax/metrics.hdf5',
    'PEPT_CnnClass4ConvKS3Lin100SoftMax' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/clustPept_noPSSM/CnnClass4ConvKS3Lin100SoftMax/metrics.hdf5',
    'ALLELE_CnnClass4ConvKS3Lin100SoftMax' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/clustAllele_noPSSM/CnnClass4ConvKS3Lin100SoftMax/metrics.hdf5',
    'SHUFF_SUmFeat_ChannExpand' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/shuffled_noPSSM_sumFeat_bn/CnnClass4ConvKS3Lin128ChannExpand/metrics.hdf5',
    'PEPT_SumFeat_ChannExpand' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/clustPept_noPSSM_sumFeat_bn/CnnClass4ConvKS3Lin128ChannExpand/metrics.hdf5', 
    'ALLELE_SumFeat_ChannExpand' : '/projects/0/einf2380/data/pMHCI/trained_models/CNN/clustAllele_noPSSM_sumFeat_bn/CnnClass4ConvKS3Lin128ChannExpand/metrics.hdf5'  
                }
showPlots = True

best_models_metrics = {}
for key, metricsfile in metricsfiles.items():
    metrics = ['acc', 'auc', 'f1', 'mcc', 'tnr', 'tpr']
    training_path = os.path.dirname(metricsfile)

    f_metrics = h5py.File(metricsfile, 'r')

    f_metrics_keys = [ x for x in f_metrics.keys() if x.startswith('epoch_')]
        
    metrics_dict = dp.get_metrics_per_epoch(f_metrics, f_metrics_keys, metrics)
    losses_dict = dp.get_losses_per_epoch(f_metrics, f_metrics_keys)

    best_model = (np.where(np.array(
        f_metrics['losses/valid'][:]) == min(f_metrics['losses/valid'][:]))[0][-1])
    
    if len(f_metrics['losses/valid'][:]) > len(f_metrics['losses/test'][:]):
        best_model = int(best_model/2)

    f_metrics.close()

    dp.export_losses(losses_dict, len(f_metrics_keys), training_path, show=showPlots, best_model=best_model)

    best_models_metrics[key] = {}
    for m in metrics:
        best_models_metrics[key][m] = dp.export_metrics(m, metrics_dict, len(f_metrics_keys), training_path, best_model=best_model, show=showPlots)

print(best_models_metrics)
```

```python
def plot_hist_comparison(title, metrics_dict, save=False):
    fig = go.Figure()

    for model_name in metrics_dict:
        x = list(metrics_dict[model_name].keys())
        y = list(metrics_dict[model_name].values())
        fig.add_trace(go.Bar(
            x=x,
            y=y,
            name=model_name,
        ))
        fig.update_layout(
            barmode="group",
            title_text=title,
        )
    fig.show()
    if save:
        plt.savefig(save)
        
plot_hist_comparison('Shuffled-peptClust-alleleClust cnn vs mlp', best_models_metrics)
```
