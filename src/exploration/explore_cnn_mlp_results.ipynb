```python
import pickle
from matplotlib import pyplot as plt
import torch
```

```python
cnn_shuffled_data = pickle.load(open("./cnn_metrics_shuffled.pkl", "rb"))
cnn_clustered_data = pickle.load(open("./cnn_metrics_clustered.pkl", "rb"))

mlp_shuffled_data = pickle.load(open("./mlp_metrics_shuffled.pkl", "rb"))
mlp_clustered_data = pickle.load(open("./mlp_metrics_clustered.pkl", "rb"))

e = range(mlp_clustered_data["test"]["accuracy"].shape[0])
fig = plt.figure(figsize=(13,18))
ax1, ax2 = fig.subplots(2)

ax1.plot(e, mlp_clustered_data["validation"]["accuracy"].mean(1),color="orange", linestyle="dashed")
ax1.plot(e, mlp_clustered_data["test"]["accuracy"].mean(1), color="orange")
ax1.plot(e, mlp_shuffled_data["test"]["accuracy"].mean(1), color="blue")
ax1.plot(e, mlp_shuffled_data["validation"]["accuracy"].mean(1), linestyle="dashed", color="blue")
ax1.set_title("MLP performances on shuffled and clustered data", fontsize=24, color="#006a90", fontweight="bold", pad=20)
ax1.set_ylabel("Accuracy", fontsize=20)
ax1.set_xlabel("Iterations over training dataset", fontsize=20)
ax1.set_ylim(.5,1)
ax1.legend(["Clustered test", "Clustered validation", "Shuffled test", "Shuffled validation"], loc="lower right", fontsize=20)
for label in (ax1.get_xticklabels() + ax1.get_yticklabels()):
    label.set_fontsize(20)

ax2.plot(e, cnn_clustered_data["test"]["accuracy"].mean(1), color="orange")
ax2.plot(e, cnn_clustered_data["validation"]["accuracy"].mean(1),color="orange", linestyle="dashed")
ax2.plot(e, cnn_shuffled_data["test"]["accuracy"].mean(1), color="blue")
ax2.plot(e, cnn_shuffled_data["validation"]["accuracy"].mean(1), linestyle="dashed", color="blue")
ax2.set_title("CNN performances on shuffled and clustered data", fontsize=24, color="#006a90", fontweight="bold", pad=20)
ax2.set_ylabel("Accuracy", fontsize=20)
ax2.set_xlabel("Iterations over training dataset", fontsize=20)
ax2.set_ylim(.5,1)
ax2.legend(["Clustered test", "Clustered validation", "Shuffled test", "Shuffled validation"], loc="lower right", fontsize=20)
for label in (ax2.get_xticklabels() + ax2.get_yticklabels()):
    label.set_fontsize(20)
```
